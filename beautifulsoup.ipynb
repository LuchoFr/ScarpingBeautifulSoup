EXPLORANDO TO SCRAPE PARA HACER SCRAPING A MULTIPLES PAGINAS

```python
url_base="http://books.toscrape.com/catalogue/page-{}.html" #Creamos url sin numero de pagina
resultado3=requests.get(url_base.format("1"))
sopa3=bs4.BeautifulSoup(resultado3.text,"lxml")
libros=sopa3.select(".product_pod") #Creo la lista con los libros
libro_1_raiting=libros[0].select(".star-rating.Three") #Los espacios vacios se cambian  con puntos si tiene 3 estrellas se guarda 
libro_1_titulo=libros[0].select("a")[1]["title"] #obtuve el titulo del libro
print(libro_1_titulo)
```

```python
#Lista de títulos con 4 o 5 estrellas
titulos_rating_alto=[]
#iteramos por páginas
for pagina in range(1,51):
    #Creamos sopa en cada pagina
    url_pagina=url_base.format(pagina)
    resultado_aux=requests.get(url_pagina)
    sopa_aux=bs4.BeautifulSoup(resultado_aux.text,"lxml")
    #Seleccionar datos de los libros
    libros=sopa_aux.select(".product_pod")
    #iterar libros
    for libro in libros:
        #chequear que tengan 4 o 5 estrellas
        if len(libro.select(".star-rating.Four"))!=0 or len(libro.select(".star-rating.Five"))!=0:
            #Guardar titulo en variable
            titulo_libro=libro.select("a")[1]["title"]
            #agregar el libro a la lista
            titulos_rating_alto.append(titulo_libro)


```

```python
for titulo in titulos_rating_alto: #Muestro la lista con los titulos
    print(titulo)
    print("")

print(titulos_rating_alto)
```
